# Import Libraries

```python
from pathlib import Path
import pandas as pd
import tqdm
tqdm.tqdm.pandas()
from fund_api import FundAPI
```

# Initialization

```python
dataset_dir = Path('output')
dataset_dir.mkdir(exist_ok=True)

fund_api = FundAPI()
```

# Fetch all fund detail from Finnomena

```python
funds = fund_api.get_funds()
fund_df = pd.DataFrame(funds)
fund_df = fund_df.loc[fund_df['sec_is_active']]
fund_df
```

```python
def retrieve_fund_detail(fund_id):
    fund_detail_dict = fund_api.get_fund(fund_id)
    daily_update_detail_dict = fund_api.get_fund_daily_update_detail(fund_id)
    
    fund_detail = pd.Series(fund_detail_dict)
    fund_detail['fund_unit_value'] = daily_update_detail_dict['value']
    fund_detail['fund_nav'] = daily_update_detail_dict['amount']
    return fund_detail

fund_detail_df = fund_df['fund_id'].progress_apply(retrieve_fund_detail)
fund_detail_df
```

```python
merge_columns = fund_detail_df.columns.difference(fund_df.columns).tolist()
index_columns = ['fund_id']

fund_df = fund_df.merge(fund_detail_df[index_columns + merge_columns], how='left', on=index_columns)
fund_df
```

```python
fund_df.to_csv(dataset_dir / 'fund.csv', index=False)
```

# Fetch portfolio data

```python
def extract_plot_element(fund_id, short_code, elements):
    fund_key = {'fund_id': fund_id, 'fund_short_code': short_code}

    return list(map(lambda element: fund_key | element, elements))


holdings = []
stock_sectors = []
asset_allocations = []
regional_exposures = []

for fund_id in tqdm.tqdm(fund_df['fund_id']):
    portfolio_dict = fund_api.get_portfolio(fund_id)
    short_code = portfolio_dict['short_code']
    if portfolio_dict['top_holdings'] is not None:
        holdings.extend(extract_plot_element(fund_id, short_code, portfolio_dict['top_holdings']['elements']))
    
    if portfolio_dict['global_stock_sector'] is not None:
        stock_sectors.extend(extract_plot_element(fund_id, short_code, portfolio_dict['global_stock_sector']['elements']))
    
    if portfolio_dict['asset_allocation'] is not None:
        asset_allocations.extend(extract_plot_element(fund_id, short_code, portfolio_dict['asset_allocation']['elements']))

    if portfolio_dict['regional_exposure'] is not None:
        regional_exposures.extend(extract_plot_element(fund_id, short_code, portfolio_dict['regional_exposure']['elements']))

drop_columns = ['color']
holding_df = pd.DataFrame(holdings).drop(columns=drop_columns)
stock_sector_df = pd.DataFrame(stock_sectors).drop(columns=drop_columns)
asset_allocation_df = pd.DataFrame(asset_allocations).drop(columns=drop_columns)
regional_exposure_df = pd.DataFrame(regional_exposures)
```

```python
holding_df
```

```python
holding_df.to_csv(dataset_dir / 'portfolio_holding.csv', index=False)
stock_sector_df.to_csv(dataset_dir / 'portfolio_stock_sector.csv', index=False)
asset_allocation_df.to_csv(dataset_dir / 'portfolio_asset_allocation.csv', index=False)
regional_exposure_df.to_csv(dataset_dir / 'portfolio_regional_exposure.csv', index=False)
```

# Fetch NAV

```python
from multiprocessing.pool import ThreadPool

```

```python
unit_values = []
navs = []
def retrieve_nav(fund_id):
    response = fund_api.get_nav(fund_id)
    short_code = response['short_code']

    nav_response = response['navs']

    if len(nav_response) == 0:
        return

    fund_nav_df = pd.DataFrame(nav_response).set_index('date')

    unit_values.append(fund_nav_df[['value']].rename(columns={'value': short_code}))
    navs.append(fund_nav_df[['amount']].rename(columns={'amount': short_code}))
```

```python
with ThreadPool(32) as pool:
    list(tqdm.tqdm(pool.imap(retrieve_nav, fund_df['fund_id']), total=len(fund_df)))
```

```python
unit_value_df = pd.concat(unit_values, axis=1).sort_index()
nav_df = pd.concat(navs, axis=1).sort_index()
```

```python
sorted_short_codes = fund_df.loc[fund_df['short_code'].isin(unit_value_df.columns), 'short_code']
unit_value_df = unit_value_df[sorted_short_codes]
nav_df = nav_df[sorted_short_codes]
```

```python
unit_value_df.to_csv(dataset_dir / 'unit_value.csv')
nav_df.to_csv(dataset_dir / 'nav.csv')
```
